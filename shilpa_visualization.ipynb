Different Visualizations

### 1. BAR GRAPH


```python
plt.bar(data['Classification'], data['PL'], color='red')
plt.title('Classification vs Petal Length')
plt.show()
```

### 2. Scatter Plots

```python
plt.figure()

plt.subplot(2,1,1)
plt.scatter(data['PL'], data['PW'],color='green')
plt.title('Petal Length vs Petal Width')
plt.xlabel('PL')
plt.ylabel('PW')
plt.show()

plt.subplot(2,1,2)
plt.scatter(data['SL'], data['SW'],color='green')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('SL')
plt.ylabel('SW')
plt.show()
```

### 3. Heatmap

```python
sns.heatmap(data.corr(), annot=True)
plt.show()
```

### 4. Box Plot

```python
sns.boxplot(x='SL',y='SW',data=data)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(data=data)
plt.show()
```
